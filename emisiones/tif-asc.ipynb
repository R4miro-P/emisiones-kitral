```python
import os
import rasterio

def tif_to_asc(input_tif, output_folder):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # Open the input TIFF
    with rasterio.open(input_tif) as src:
        num_bands = src.count
        print(f"Number of bands: {num_bands}")
        
        # Copy the profile and update it for ASCII grid output.
        # We force count to 1 because each ASC file contains one band.
        profile = src.profile.copy()
        profile.update(driver='AAIGrid', dtype=rasterio.float32, count=1)
        
        # Loop over each band and write to an ASC file
        for band in range(1, num_bands + 1):
            band_array = src.read(band)  # Read the current band
            output_path = os.path.join(output_folder, f"band_{band}.asc")
            with rasterio.open(output_path, 'w', **profile) as dst:
                dst.write(band_array.astype(rasterio.float32), 1)
            print(f"Band {band} written to {output_path}")

# Example usage:
input_tif = "/home/ramiro/Emisiones/50sim/BurnProbability_extremo.tif"          # Replace with your TIFF file path
output_folder = "/home/ramiro/Emisiones/50sim/"    # Replace with your desired output folder
tif_to_asc(input_tif, output_folder)
```

```python
def convert_tif_to_asc(input_tif_path, output_asc_path):
    
    #Convierte un GeoTIFF en un Arc/Info ASCII Grid.

    with rasterio.open(input_tif_path) as src:
        data = src.read(1)
        nodata = src.nodata
        if nodata is not None:
            mask = data == nodata
            data = np.where(mask, 0, data)
        
        profile = src.profile.copy()
        profile.update(driver='AAIGrid', dtype=rasterio.float32, nodata=nodata if nodata is not None else 0, count=1)
    
    with rasterio.open(output_asc_path, 'w', **profile) as dst:
        dst.write(data.astype(np.float32), 1)
    
    print(f"Converted {input_tif_path} to ASCII grid: {output_asc_path}")
```

```python
import os
import numpy as np
import rasterio

input_folder = "/home/ramiro/Emisiones/50sim/CFB_manual_extremo/"
output_file = "/home/ramiro/Emisiones/50sim/cicatriz_total_copa_extremo.asc" 

def count_valid_pixels(input_folder, output_file):
    # List all .asc files in the input folder.
    files = [f for f in os.listdir(input_folder) if f.endswith('.asc')]
    if not files:
        raise FileNotFoundError("No .asc files found in the specified folder.")
    
    # Open the first file to obtain the profile and dimensions.
    first_file = os.path.join(input_folder, files[0])
    with rasterio.open(first_file) as src:
        profile = src.profile.copy()
        data = src.read(1)
    
    # Initialize an array to accumulate the count.
    count_array = np.zeros_like(data, dtype=np.float32)
    
    # Loop over each file and accumulate counts.
    for idx, file in enumerate(files):
        print(f"Processing file {idx+1}/{len(files)}: {file}")
        file_path = os.path.join(input_folder, file)
        with rasterio.open(file_path) as src:
            data = src.read(1)
        # Replace negative values with 0.
        data[data < 0] = 0
        # For each pixel, if data is nonzero, add 1.
        count_array += (data != 0).astype(np.float32)
    
    # Update the profile for ASCII grid output.
    profile.update(driver="AAIGrid", dtype=rasterio.float32, count=1)
    
    # Write the output ASC file with the count array.
    with rasterio.open(output_file, "w", **profile) as dst:
        dst.write(count_array, 1)
    
    print(f"Output written to: {output_file}")

# Run the function.
count_valid_pixels(input_folder, output_file)
```

```python
import os
import numpy as np
import rasterio

def update_cfb_with_crown_fraction(cfb_file, crown_file, output_file):
    """
    Reads a CFB file and its corresponding crown_fraction_burned file.
    For each pixel, if the crown_fraction_burned value is 0, the corresponding CFB pixel is set to 0.
    The updated array is then written to output_file.
    """
    with rasterio.open(cfb_file) as src:
        cfb_data = src.read(1)
        profile = src.profile.copy()
        
    with rasterio.open(crown_file) as src:
        crown_data = src.read(1)
    
    # Set CFB pixel to 0 wherever the crown fraction is 0.
    updated_cfb = np.where(crown_data == 0, 0, cfb_data)
    
    with rasterio.open(output_file, "w", **profile) as dst:
        dst.write(updated_cfb, 1)
    
    return output_file

def process_cfb_files(cfb_folder, crown_folder, output_folder):
    """
    For each ASC file in the CFB folder, this function:
      1. Extracts the band identifier (e.g., from "CFB_band1.asc" it extracts "band1.asc").
      2. Finds the corresponding crown_fraction_burned file in crown_folder.
      3. Updates the CFB file so that wherever the crown_fraction_burned value is 0, 
         the CFB pixel is set to 0.
      4. Saves the modified CFB file into output_folder.
    """
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # List all ASC files in the CFB folder.
    cfb_files = [f for f in os.listdir(cfb_folder) if f.lower().endswith('.asc')]
    
    for file in cfb_files:
        cfb_file_path = os.path.join(cfb_folder, file)
        
        # Assume CFB files are named "CFB_bandX.asc" and we want "bandX.asc" for matching.
        if file.startswith("crown_fraction_burned_"):
            band_name = file[len("crown_fraction_burned_"):]
        else:
            band_name = file  # If there's no prefix, assume it's already the band name.
        
        crown_file_path = os.path.join(crown_folder, band_name)
        
        if not os.path.exists(crown_file_path):
            print(f"Corresponding CFB file for {file} not found. Skipping...")
            continue
        
        output_file_path = os.path.join(output_folder, file)
        update_cfb_with_crown_fraction(cfb_file_path, crown_file_path, output_file_path)
        print(f"Processed {file}: updated CFB_manual saved to {output_file_path}")

# Example usage:
cfb_folder = "/home/ramiro/Emisiones/50sim/CFB_manual"
crown_folder = "/home/ramiro/Emisiones/50sim/CFB"
output_folder = "/home/ramiro/Emisiones/50sim/modified_CFB"

process_cfb_files(cfb_folder, crown_folder, output_folder)

```

```python
import os
import rasterio
import numpy as np

def count_nonzero_cells(raster_path):
    """
    Counts the number of nonzero cells in a raster using Rasterio.
    It reads the first band as a masked array (ignoring nodata) and fills masked values with 0.
    """
    with rasterio.open(raster_path) as src:
        data = src.read(1, masked=True)
        # Fill masked values with 0 and count non-zero cells
        nonzero_count = np.count_nonzero(data.filled(0))
    return nonzero_count

def compute_counts_ratio(sfb_folder, cfb_folder):
    """
    For each file in the SFB folder (with names like SFB_band31.asc),
    this function looks for a matching file in the CFB folder (e.g., CFB_band31.asc),
    computes the count of nonzero cells in each, and returns a dictionary with the ratio:
       ratio = count_CFB / count_SFB.
    """
    # List SFB and CFB files (assuming they end with ".asc")
    sfb_files = sorted([f for f in os.listdir(sfb_folder) if f.lower().endswith('.asc') and f.startswith("surface_fraction_burned")])
    cfb_files = sorted([f for f in os.listdir(cfb_folder) if f.lower().endswith('.asc') and f.startswith("crown_fraction_burned")])
    
    ratios = {}
    
    # Loop over each SFB file and compute ratio with corresponding CFB file.
    for sfb_file in sfb_files:
        # Extract the band identifier; e.g., from "SFB_band31.asc" extract "band31.asc"
        band_identifier = sfb_file[len("surface_fraction_burned_"):]
        corresponding_cfb_file = "crown_fraction_burned_" + band_identifier
        
        sfb_path = os.path.join(sfb_folder, sfb_file)
        cfb_path = os.path.join(cfb_folder, corresponding_cfb_file)
        
        if not os.path.exists(cfb_path):
            print(f"CFB file for {sfb_file} not found. Skipping...")
            continue
        
        count_sfb = count_nonzero_cells(sfb_path)
        count_cfb = count_nonzero_cells(cfb_path)
        
        if count_sfb == 0:
            ratio = np.nan
            print(f"For band {band_identifier}: count_SFB is 0. Setting ratio to NaN.")
        else:
            ratio = count_cfb / count_sfb
        
        ratios[band_identifier] = ratio
        print(f"Band {band_identifier}: CFB count = {count_cfb}, SFB count = {count_sfb}, Ratio = {ratio}")
    
    return ratios

# Update these paths to your actual folder locations:
sfb_folder = "/home/ramiro/Emisiones/50sim/SFB"
cfb_folder = "/home/ramiro/Emisiones/50sim/modified_CFB"

ratios = compute_counts_ratio(sfb_folder, cfb_folder)
print("\nFinal Ratios (CFB/SFB):")
for band, ratio in ratios.items():
    print(f"{band}: {ratio}")


```
